```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sp
import scipy.fftpack
```

```python
fs=10000
```

```python
mean = 0
std = 1
num_samples = 500000
white_noise = np.random.normal(mean, std, size=num_samples)
```

```python
plt.plot(white_noise)
```

```python
white_noise_fft = np.fft.fft(white_noise)

print(white_noise_fft.shape)
```

```python
freq = np.fft.fftfreq(white_noise_fft.shape[-1], 1/fs)
```

```python
plt.plot(freq)
```

```python
i = freq > 0
```

```python
plt.plot(freq[i], white_noise_fft[i])
```

```python
print(white_noise_fft.shape)
```

```python
white_noise_recover = np.fft.ifft(white_noise_fft)
```

```python
plt.plot(white_noise_recover)
```
